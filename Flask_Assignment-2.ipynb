**Q1. Explain GET and POST methods.**

GET and POST are two commonly used methods in the HTTP protocol for sending data between a client and a server. 

The **GET method** is used to retrieve data from a server. When a client sends a GET request to a server, it is asking the server to send back a representation of a specific resource.

The **POST method** is used to send data to the server to be processed or stored. When a client sends a POST request, the data is included in the body of the HTTP request rather than the URL.

**Q2. Why is request used in Flask?**

In Flask, the request object is used to access incoming client data from an HTTP request. Moreover, it also provides convenient access to various parts of the request, such as the headers, URL parameters, query parameters, form data, and uploaded files. 

**Q3. Why is redirect() used in Flask?**

redirect() function in Flask is used to redirect the client to a different URL. It is a convenient way to instruct the client's web browser to navigate to a different locations.

**Q4. What are templates in Flask? Why is the render_template() function used?**

Templates in Flask are files that contain HTML markup mixed with placeholders for dynamic content. 

The render_template() function in Flask is used to render and display templates in the web browser.

These are used:

1. The render_template() function is used to process the template file, replacing template variables with actual values.

2. This function is responsible for combining the template and the data, allowing you to focus on the application logic separately.

3. It also handles the rendering and inheritance process, making it easy to create modular and reusable templates.
